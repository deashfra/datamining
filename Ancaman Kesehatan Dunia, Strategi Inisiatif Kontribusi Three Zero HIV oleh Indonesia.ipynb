** K means clustering Inertia/Elbow method**

1. klustering provinsi berdasarkan pendidikan akhir

```python
df_select_1 = df[['SD/SEDERAJAT','SMP/SEDERAJAT','SMA/SEDERAJAT']]
```

```python
df_select_1.describe()
```

```python
#normalisasi
from sklearn.preprocessing import StandardScaler, MinMaxScaler

norm_col = ['SD/SEDERAJAT','SMP/SEDERAJAT','SMA/SEDERAJAT']

for col in norm_col:
  df_select_1[col + '_NORM'] = MinMaxScaler().fit_transform(df_select_1[col].values.reshape(len(df_select_1), 1))
```

```python
df_select_1.describe().apply(lambda x: x.apply('{0:.5f}'.format))
```

```python
df_select_std_1 = df_select_1.copy()
df_select_std_1 = df_select_std_1[['SD/SEDERAJAT_NORM',	'SMP/SEDERAJAT_NORM',	'SMA/SEDERAJAT_NORM',]]
```

```python
 from sklearn.cluster import KMeans
inertia = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_select_std_1)
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(20, 10))

sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
plt.show()
```

```python
kmeans_1 = KMeans(n_clusters=3, random_state=0).fit(df_select_std_1)
```

```python
clusters_pendidikan = kmeans_1.labels_
```

```python
df2 = df.copy()
df2['cluster_pendidikan'] = clusters_pendidikan
```

```python
df2
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(df_select_std_1)
pcs = pca.transform(df_select_std_1)

df_pca_1 = pd.DataFrame(data = pcs, columns = ['PC 1', 'PC 2'])
df_pca_1.head()
```

```python
df_pca_1['clusters'] = clusters_pendidikan
```

```python
df_pca_1
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.scatterplot(
    x='PC 1', y='PC 2',
    hue='clusters',
    linestyle='--',
    data=df_pca_1,
    s=160,
    ax=ax
)


plt.show()
```

0 : pendidikan cukup tinggi
1 : pendidikan menengah
2 :Pendidikan rendah

# Grafik tren kasus total hiv di indonesia

```python
#Time series indonesia
series = pd.read_excel('DATA INFODATIN 2020.xlsx')
series.head()
```

```python
series
```

```python
def plot_df(series, x, y, title="", xlabel='Tahun', ylabel='Jumlah Kasus HIV', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()


plot_df(series, x=series['Tahun'], y=series['Jumlah Kasus HIV'], title='Trend')
```

```python
# Analisis tren menggunakan regresi linier
from sklearn.linear_model import LinearRegression

X = series['Tahun'].values.reshape(-1, 1)
y = series['Jumlah Kasus HIV'].values.reshape(-1, 1)

regression = LinearRegression()
regression.fit(X, y)

trend = regression.predict(X)
```

```python
plt.plot(series['Tahun'], series['Jumlah Kasus HIV'], label='Observasi')
plt.plot(series['Tahun'], trend, label='Tren')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kasus HIV')
plt.title('Jumlah Kasus HIV per Tahun')
plt.legend()
plt.show()
```

Grafik menunjukan adanya tren positif yang artinya total kasus akan meningkat seiring dengan berjalannya waktu.
